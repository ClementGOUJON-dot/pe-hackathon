```python
# Compter les découvertes par année et méthode

data = pd.read_csv('cleaned_data.csv')

data_counts = data.groupby(['sy_dist', 'discoverymethod']).size().reset_index(name='discoveries')

# Calculer le cumul des découvertes pour chaque méthode
data_counts['cumulative_discoveries'] = data_counts.groupby('discoverymethod')['discoveries'].cumsum()

# Extraire la dernière distance pour chaque méthode
#last_year_counts = data_counts[data_counts['sy_dist'] == data_counts['sy_dist'].max()]

# Trier les méthodes par découvertes cumulées finales
#sorted_methods = last_year_counts.sort_values('cumulative_discoveries')['discoverymethod'].tolist()

# Initialiser le graphique avec les couleurs dans l'ordre trié
plt.figure(figsize=(14, 8))
sns.histplot(data=data_counts, x="sy_dist", weights="cumulative_discoveries", hue="discoverymethod", multiple="stack", palette="Set2", bins=30)

# Ajouter les labels et un titre

plt.yscale('log')
plt.xlabel("Distance")
plt.ylabel("Nombre Cumulé Total de Planètes Découvertes")
plt.title("Évolution Cumulative des Découvertes de Planètes par Méthode de Découverte et par distance (Ordre Croissant à la Fin)")

# Afficher le graphique
plt.show()
```
